### Research Question 1: Is the revenue related to the budget?


> Use this, and more code cells, to explore your data. Don't forget to add
>   Markdown cells to document your observations and findings.


```python
plt.rcParams['figure.figsize']=(10,8)
dffinal.plot(x= 'budget', y='revenue', kind='scatter');
plt.title('Scatter Plot showing the relationship between budget and revenue')
plt.xlabel('Budget (100 Million Dollars)')
plt.ylabel('Revenue (Billion Dollars)')
plt.show()
```

The scatter plot indicates that to some extent as the budget increases the revenue also increases. However, this conclusion might be premature because there are many points where the budget was extremely high yet the revenue did not come parallel to the expectation.
To have a better view about the relationship between budget and revenue I will need to go deeper in understanding the budget parameter.

```python
dffinal.describe()
```

```python
dffinal['budget'].describe()
```

```python
plt.rcParams['figure.figsize']=(10,6)

dffinal['budget'].plot.box(vert = False)

plt.title('Box Plot summarizing Budget')
plt.xlabel('Budget (100 Million Dollars)')

plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,6)
dffinal['budget'].plot.hist()

plt.title('Histogram summarizing Budget')
plt.xlabel('Budget (100 Million Dollars)')

plt.show()
```

```python
dffinal['budget'].head(10)
```

The mean and the very high standard deviaion in addition to the histogram and the boxplot indicate that the budget might not be normaly distributed. For better intuition about the role of the budget in predecting the revenue I will split the data to low budget movies (<mean budget) and high budget movies (below mean budget) and high budget movies (costed an above mean budget).

```python
# get the mean budget (Mean = dffinal.budget.mean())

# Split: Low budget movies
df_LowBud = dffinal[dffinal.budget < (dffinal.budget.mean())]
```

```python
plt.rcParams['figure.figsize']=(12,8)

df_LowBud.plot(x= 'budget', y='revenue', kind='scatter');

plt.title('Scatter Plot showing the relationship between budget and revenue for low budget movies')
plt.xlabel('Budget (100 Million Dollars)')
plt.ylabel('Revenue (Billion Dollars)')
plt.show()
```

```python
# Split: Low budget movies
df_HighBud = dffinal[dffinal.budget > (dffinal.budget.mean())]
```

```python
plt.rcParams['figure.figsize']=(12,8)
df_HighBud.plot(x= 'budget', y='revenue', kind='scatter');

plt.title('Scatter Plot showing the relationship between budget and revenue for high budget movies')
plt.xlabel('Budget (100 Million Dollars)')
plt.ylabel('Revenue (Billion Dollars)')
plt.show()
```

Now it is more obvious that as the budget increases the revenue increases. This effect is more obvious for movies with relatively high budgets (> mean budgets in the market).

### Research Question 3: which movie genres were more frequent?

```python
dffinal['1st_Genre'].unique()
```

```python
Genres1 = dffinal['1st_Genre']
Genres2 = dffinal['2nd_Genre']
Genres3 = dffinal['3rd_Genre']
Genres4 = dffinal['4th_Genre']
Genres5 = dffinal['5th_Genre']


Genres = Genres1.append(Genres2)
Genres = Genres.append(Genres3)
Genres = Genres.append(Genres4)
Genres = Genres.append(Genres5)



from collections import Counter
#source: https://stackoverflow.com/questions/2600191/how-to-count-the-occurrences-of-a-list-item

x = Counter(Genres)
print(x)
```

```python
#Sorting the file
del x['Unknown']
x.most_common()

#Source: https://stackoverflow.com/questions/20950650/how-to-sort-counter-by-value-python
```

```python
z = x.keys()
y = x.values()

plt.rcParams['figure.figsize']=(25,6)

plt.bar(z,y, align='center', );
plt.title('Bar chart showing the frequencies of each movie genre through the years')
plt.show()
```
